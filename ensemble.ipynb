```python
#Eval Brixia
import torch
from torchvision import transforms
import cxr_dataset as CXR

import eval_model as E

import numpy

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
PATH_TO_IMAGES = "/content/gdrive/MyDrive/ColabNotebooks/brixia/DB"
MODEL_PATH = '/content/gdrive/MyDrive/ColabNotebooks/brixia/results/regression_checkpoint_best'
MODEL_PATH2 = '/content/gdrive/MyDrive/ColabNotebooks/brixia/2results/regression_checkpoint_best'
MODEL_PATH3 = '/content/gdrive/MyDrive/ColabNotebooks/brixia/3results/regression_checkpoint_best'
MODEL_PATH4 = '/content/gdrive/MyDrive/ColabNotebooks/brixia/4results/regression_checkpoint_best'
MODEL_PATH5 = '/content/gdrive/MyDrive/ColabNotebooks/brixia/5results/regression_checkpoint_best'


# # use imagenet mean,std for normalization
mean = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]

data_transforms = {
    'train': transforms.Compose([
        transforms.Resize(224),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(mean, std)
    ]),
    'val': transforms.Compose([
        transforms.Resize(224),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(mean, std)
    ]),
}

# create dataloader
dataset = CXR.CXRDataset(
    path_to_images=PATH_TO_IMAGES,
    fold='test',
    transform=data_transforms['val'],
    fine_tune=True,
    regression=True)
dataloader = torch.utils.data.DataLoader(
    dataset, 32, shuffle=False, num_workers=8)

saved_model = torch.load(MODEL_PATH)
model = saved_model['model']
model.to(device)
del saved_model

saved_model2 = torch.load(MODEL_PATH2)
model2 = saved_model2['model']
model2.to(device)
del saved_model2

saved_model3 = torch.load(MODEL_PATH3)
model3 = saved_model3['model']
model3.to(device)
del saved_model3

saved_model4 = torch.load(MODEL_PATH4)
model4 = saved_model4['model']
model4.to(device)
del saved_model4

saved_model5 = torch.load(MODEL_PATH5)
model5 = saved_model5['model']
model5.to(device)
del saved_model5

mae, ground_truths, preds = E.evaluate_mae(dataloader, model)
print('MAE : ', mae)
print('True_df : ',ground_truths)
print('Pred_df : ', preds)

mae2, ground_truths2, preds2 = E.evaluate_mae(dataloader, model2)
print('MAE2 : ', mae2)
print('True_df2 : ',ground_truths2)
print('Pred_df2 : ', preds2)

mae3, ground_truths3, preds3 = E.evaluate_mae(dataloader, model3)
print('MAE3 : ', mae3)
print('True_df3 : ',ground_truths3)
print('Pred_df3 : ', preds3)

mae4, ground_truths4, preds4 = E.evaluate_mae(dataloader, model4)
print('MAE4 : ', mae4)
print('True_df4 : ',ground_truths4)
print('Pred_df4 : ', preds4)

mae5, ground_truths5, preds5 = E.evaluate_mae(dataloader, model5)
print('MAE5 : ', mae5)
print('True_df5 : ',ground_truths5)
print('Pred_df5 : ', preds5)

maes = []
maes.append(mae)
maes.extend([mae2, mae3, mae4, mae5])
print(maes)
mae_mean = numpy.mean(maes)
print("성능 : ", mae_mean)
```

```python
import pandas as pd

## 일부 출력
# pd.options.display.max_rows = 30
# pd.options.display.max_columns = 5

# 예측값 table(각 방법별 예측값 + var(uncertainty) + mean(예측값 평균))
preds_outer = pd.merge(preds, preds2, how="outer", on="Image Index")
preds_outer = pd.merge(preds_outer, preds3, how="outer", on="Image Index")
preds_outer = pd.merge(preds_outer, preds4, how="outer", on="Image Index")
preds_outer = pd.merge(preds_outer, preds5, how="outer", on="Image Index")

# print(preds_outer) 
# print(type(preds_outer))
print(preds_outer.columns)

# 각 image 별 variance, mean 추가
preds_outer['Var']= preds_outer[['pred_score_x','pred_score_y','pred_score_x','pred_score_y','pred_score']].var(axis=1).values
preds_outer['Mean']= preds_outer[['pred_score_x','pred_score_y','pred_score_x','pred_score_y','pred_score']].mean(axis=1).values

print(preds_outer)
```

```python
var_array = preds_outer['Var']
print(var_array)
sort_var = var_array.sort_values(ascending=False) # 내림차순
print("\n불확실성 높은 순 : \n",sort_var)

# test Data : 총 207개
# 95% --> 약 10개
# 90% --> 약 20개
# 85% --> 약 31개
drop_table = preds_outer.loc[:,['Image Index','Var']]
print(drop_table)
sort_drop = drop_table.sort_values('Var',ascending=False) # 내림차순
print("정렬\n")
print(sort_drop)

# 95%
uncertainty_up95 = sort_drop.iloc[0:10]
print(uncertainty_up95)
print(len(uncertainty_up95))
image_95 = uncertainty_up95.iloc[:,0]
print(image_95)

# 90%
uncertainty_up90 = sort_drop.iloc[0:20]
print(uncertainty_up90)
print(len(uncertainty_up90))
drop_90 = uncertainty_up90.iloc[:,0]
print("90% (drop 20) : \n", drop_90)

# 85%
uncertainty_up85 = sort_drop.iloc[0:31]
print(uncertainty_up85)
print(len(uncertainty_up85))
drop_85 = uncertainty_up85.iloc[:,0]
print("85% (drop 31) : \n", drop_85)
```

```python
#Eval Brixia
import torch
from torchvision import transforms
import cxr_dataset as CXR

import eval_model as E

import numpy

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
PATH_TO_IMAGES = "/content/gdrive/MyDrive/ColabNotebooks/brixia/DB/ensemble_per95"
MODEL_PATH = '/content/gdrive/MyDrive/ColabNotebooks/brixia/results/regression_checkpoint_best'
MODEL_PATH2 = '/content/gdrive/MyDrive/ColabNotebooks/brixia/2results/regression_checkpoint_best'
MODEL_PATH3 = '/content/gdrive/MyDrive/ColabNotebooks/brixia/3results/regression_checkpoint_best'
MODEL_PATH4 = '/content/gdrive/MyDrive/ColabNotebooks/brixia/4results/regression_checkpoint_best'
MODEL_PATH5 = '/content/gdrive/MyDrive/ColabNotebooks/brixia/5results/regression_checkpoint_best'


# # use imagenet mean,std for normalization
mean = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]

data_transforms = {
    'train': transforms.Compose([
        transforms.Resize(224),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(mean, std)
    ]),
    'val': transforms.Compose([
        transforms.Resize(224),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(mean, std)
    ]),
}

# create dataloader
dataset = CXR.CXRDataset(
    path_to_images=PATH_TO_IMAGES,
    fold='test',
    transform=data_transforms['val'],
    fine_tune=True,
    regression=True)
dataloader = torch.utils.data.DataLoader(
    dataset, 32, shuffle=False, num_workers=8)

saved_model = torch.load(MODEL_PATH)
model = saved_model['model']
model.to(device)
del saved_model

saved_model2 = torch.load(MODEL_PATH2)
model2 = saved_model2['model']
model2.to(device)
del saved_model2

saved_model3 = torch.load(MODEL_PATH3)
model3 = saved_model3['model']
model3.to(device)
del saved_model3

saved_model4 = torch.load(MODEL_PATH4)
model4 = saved_model4['model']
model4.to(device)
del saved_model4

saved_model5 = torch.load(MODEL_PATH5)
model5 = saved_model5['model']
model5.to(device)
del saved_model5

mae, ground_truths, preds = E.evaluate_mae(dataloader, model)
print('MAE : ', mae)
print('True_df : ',ground_truths)
print('Pred_df : ', preds)

mae2, ground_truths2, preds2 = E.evaluate_mae(dataloader, model2)
print('MAE2 : ', mae2)
print('True_df2 : ',ground_truths2)
print('Pred_df2 : ', preds2)

mae3, ground_truths3, preds3 = E.evaluate_mae(dataloader, model3)
print('MAE3 : ', mae3)
print('True_df3 : ',ground_truths3)
print('Pred_df3 : ', preds3)

mae4, ground_truths4, preds4 = E.evaluate_mae(dataloader, model4)
print('MAE4 : ', mae4)
print('True_df4 : ',ground_truths4)
print('Pred_df4 : ', preds4)

mae5, ground_truths5, preds5 = E.evaluate_mae(dataloader, model5)
print('MAE5 : ', mae5)
print('True_df5 : ',ground_truths5)
print('Pred_df5 : ', preds5)

maes = []
maes.append(mae)
maes.extend([mae2, mae3, mae4, mae5])
print(maes)
mae_mean = numpy.mean(maes)
print("성능 : ", mae_mean)
```
